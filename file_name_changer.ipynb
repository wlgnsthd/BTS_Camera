```python
import os
#시작 번호 수정
n = 201
```

```python
#대상폴더 수정
Dest = 'C:/Users/JHS_WIN/Desktop/2_200/'
```

```python
#대상 파일확장자 수정
fileEx = r'.jpg'
jpg_list = [file for file in os.listdir(Dest) if file.endswith(fileEx)]
```

```python
jpg_list
```

```python
for idx in jpg_list:
    src_file = Dest + idx
    #수정 파일명
    output = Dest + 'LadderTruck ('+str(n) +')'+'.jpg'
    print('{} --> {}'.format(src_file, output))
    os.rename(src_file, output)
    n = n + 1
```
